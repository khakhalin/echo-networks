```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import networkx as nx
import argraph as ag
```

```python
# Naive test of the correlation coefficient transmission curve

signal = np.random.normal(size=[1,100])
c = [0]*11
for j in range(0,11):
    level = j/10
    c[j] = 0
    for i in range(0,100):
        temp = np.corrcoef(signal,signal*level+(1-level)*np.random.normal(size=np.shape(signal)))
        c[j] += temp[0,1]
    c[j] = c[j]/100
    
plt.plot(np.arange(0,11,1)/10,c)
plt.grid(True)
```

```python
# A matrix of zeros with 10 randomly placed ones
nNodes = 50
Arandom = np.zeros(nNodes*nNodes)
Arandom[np.random.permutation(nNodes*nNodes)[1:10]] = 1
Arandom = Arandom.reshape([nNodes,nNodes])
plt.imshow(Arandom, cmap="Greys")
```

```python
# Dynamic plot test
%matplotlib notebook 
# This is needed to make the plot dynamic
# (otherwise only the last plot gets rendered)

fig,ax = plt.subplots(1,1)

nNodes = 50
Arandom = np.zeros(nNodes*nNodes)
for i in range(1,100):
    Arandom[np.random.permutation(nNodes*nNodes)[1:2]] = 1
    ax.clear()
    ax.imshow(Arandom.reshape([nNodes,nNodes]), cmap="Greys")
    fig.canvas.draw()
    time.sleep(0.001) # It's slow enough to work as is, but just in case
```
