# Visual test of sampling in our data generator

```python
import numpy as np
import matplotlib.pyplot as plt
from echo import esn
```

```python
steps = [0.01, 0.02, 0.05, 0.1, 0.2, 0.5, 1, 2, 5]

data = esn.Data.create_source('lorenz')

for i in range(len(steps)):    
    x, y = data.generate(200, sampling_step = steps[i])
    plt.figure(figsize=(10,2))
    plt.plot(y)
```
