# Top-level simple experiment with the new version of echo

```python
import numpy as np
import matplotlib.pyplot as plt
from echo import esn
```

```python
data = esn.Data.create_source('lorenz')
x,y = data.generate(1000, sampling_step=0.021)
plt.figure(figsize=(12,3))
plt.plot(x);
plt.plot(y);
```

```python
print(f"X: mean={np.mean(x):2.2f}, std={np.std(x):2.2f}")
print(f"Y: mean={np.mean(y):2.2f}, std={np.std(y):2.2f}")
```

```python
model = esn.Reservoir(100)
print(model)
w = model.fit(x, y)
plt.plot(w);
```

```python
z = model.predict(x)
plt.figure(figsize=(12,3))
plt.plot(y, 'gray')
plt.plot(z);
```
